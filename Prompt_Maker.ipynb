# Prompt maker


```python
import pandas as pd
from openpyxl import Workbook

# Function to generate a specified number of prompts and include test labels
def generate_prompts(train_filepath, test_filepath, num_prompts, num_rows_per_label):
    # Load the datasets from the files
    train = pd.read_csv(train_filepath)
    test = pd.read_csv(test_filepath)

    # Sort datasets by 'label' column
    train = train.sort_values(by='label')
    test = test.sort_values(by='label')

    # Initialize an empty list to hold the prompts and test labels
    prompts = []
    test_labels = []

    # Generate the specified number of prompts
    for _ in range(num_prompts):
        # Initialize an empty dictionary to hold selected rows
        selected_rows = {}

        # Initialize an empty string to hold the current prompt
        prompt = ""

        # Loop over each label
        for label in ['positive', 'negative', 'no impact']:
            # Get all rows with the current label
            label_rows = train[train['label'] == label]

            # Randomly select the specified number of rows
            selected_rows[label] = label_rows.sample(num_rows_per_label)

            # Drop the selected rows from the training dataset
            train = train.drop(selected_rows[label].index)

        # Randomly select a row from the testing dataset
        test_row = test.sample(1)

        # Add selected rows and question to the current prompt
        for label, rows in selected_rows.items():
            prompt += f"Your task is to do the classification. I will provide you with data points of label 1, label 2 and label 3. Your task is to learn from them.\n"
            prompt += f"Label {label}:\n\n"
            for i, row in enumerate(rows.itertuples(), 1):
                prompt += f"{i}. {row.sentence}\n"
            prompt += "\n"

        prompt += f"Can you classify the next sentence to the above labels? Please do your best and try to choose one label.\n"
        prompt += f"Sentence: {test_row.iloc[0]['sentence']}\n"

        # Add the current prompt to the list of prompts
        prompts.append(prompt)

        # Add the label of the test row to the list of test labels
        test_labels.append(test_row.iloc[0]['label'])

    # Save the updated training dataset to the file
    train.to_csv(train_filepath, index=False)

    # Return the list of prompts and the list of test labels
    return prompts, test_labels

# Call the function to generate 50 prompts with 3 rows per label
prompts, test_labels = generate_prompts('/content/data/f_train.csv', '/content/data/f_test.csv', 50, 5)

# Create a new workbook and select the active sheet
wb = Workbook()
ws = wb.active

# Write each prompt and corresponding test label to separate cells
for i, (prompt, test_label) in enumerate(zip(prompts, test_labels), 1):
    ws[f'A{i}'] = prompt
    ws[f'B{i}'] = test_label

# Save the workbook to a file
wb.save('/content/results/promptsf_to_p.xlsx')
```

```python
import pandas as pd
from typing import List, Tuple

# Function to generate a specified number of prompts and include test labels
def generate_prompts(train_filepath: str, test_filepath: str, num_prompts: int, num_rows_per_label: int) -> Tuple[List[str], List[str]]:
    # Load the datasets from the files
    train = pd.read_csv(train_filepath)
    test = pd.read_csv(test_filepath)

    # Sort datasets by 'label' column
    train = train.sort_values(by='label')
    test = test.sort_values(by='label')

    # Initialize an empty list to hold the prompts and test labels
    prompts = []
    test_labels = []

    # Label dictionary to map labels to numbers
    label_dict = {'positive': 1, 'negative': 2, 'no impact': 3}

    # Generate the specified number of prompts
    for _ in range(num_prompts):
        # Initialize an empty dictionary to hold selected rows
        selected_rows = {}

        # Initialize an empty string to hold the current prompt
        prompt = ""

        # Loop over each label
        for label in ['positive', 'negative', 'no impact']:
            # Get all rows with the current label
            label_rows = train[train['label'] == label]

            # Randomly select the specified number of rows
            selected_rows[label] = label_rows.sample(num_rows_per_label)

            # Drop the selected rows from the training dataset
            train = train.drop(selected_rows[label].index)

        # Randomly select a row from the testing dataset
        test_row = test.sample(1)


        prompt += f"Your task is to do the classification. I will provide you with data points of label 1, label 2 and label 3. Your task is to learn from them.\n"

        # Add selected rows and question to the current prompt
        for label, rows in selected_rows.items():
            prompt += f"Label {label_dict[label]}:\n\n"
            for i, row in enumerate(rows.itertuples(), 1):
                prompt += f"{i}. {row.sentence}\n"
            prompt += "\n"

        prompt += f"Can you classify the next sentence to the above labels? Please do your best and try to choose one label.\n"
        prompt += f"Sentence: {test_row.iloc[0]['sentence']}\n"

        # Add the current prompt to the list of prompts
        prompts.append(prompt)

        # Add the label of the test row to the list of test labels
        test_labels.append(test_row.iloc[0]['label'])

    # Save the updated training dataset to the file
    train.to_csv(train_filepath, index=False)

    # Return the list of prompts and the list of test labels
    return prompts, test_labels

# Call the function to generate 50 prompts with 3 rows per label
prompts, test_labels = generate_prompts('/content/data/f_train.csv', '/content/data/poem_train.csv', 50, 5)

# Create a DataFrame with the prompts and labels
df_results = pd.DataFrame({
    "Prompt": prompts,
    "Label": test_labels
})

# Save the DataFrame to an Excel file
df_results.to_excel("/content/data/promptsf_to_p.xlsx", index=False)
```
